```python
import collections
graph = {
    0: [1, 2, 3],
    1: [ 4, 5],
    2: [6],
    3: [7],
    4: [],
    5: [],
    6: [],
    7: []
}

def bfs(graph, start):
    visited = []
    queue = collections.deque([start])
    while queue:
        vertex = queue.popleft()
        visited.append(vertex)
        
        for neighbor in graph[vertex]:
            queue.append(neighbor)
            
    return visited

def dfs(graph, node, visited=None):
        if visited is None:
            visited = []
        
        visited.append(node)
        
        for neighbor in graph[node]:
            dfs(graph, neighbor, visited)
            
        return visited
print("BFS Output:", bfs(graph, 0))
print("DFS Output:", dfs(graph, 0))
```

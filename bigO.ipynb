```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Input sizes to compare
input_sizes = [5, 10, 15]

# คำนวณค่าของ Big O สำหรับแต่ละ input size
labels = ['O(1)', 'O(log n)', 'O(n)', 'O(n log n)', 'O(n²)', 'O(2ⁿ)', 'O(n!)']
data = {n: [
    1,                          # O(1)
    math.log2(n),              # O(log n)
    n,                         # O(n)
    n * math.log2(n),          # O(n log n)
    n**2,                      # O(n^2)
    2**n,                      # O(2^n)
    math.factorial(n)          # O(n!)
] for n in input_sizes}

# ตำแหน่งของแท่งแต่ละกลุ่ม
x = np.arange(len(labels))
width = 0.25

# สร้าง bar graph
fig, ax = plt.subplots(figsize=(12, 6))
bars = []

for i, n in enumerate(input_sizes):
    bars.append(
        ax.bar(x + i * width, data[n], width, label=f"n = {n}")
    )

# กำหนดชื่อแกน x และตำแหน่ง
ax.set_xlabel('Big-O Complexity')
ax.set_ylabel('Operation Count (approx)')
ax.set_title('Big-O Time Complexity (Bar Graph)')
ax.set_xticks(x + width)
ax.set_xticklabels(labels)
ax.legend()
ax.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

```

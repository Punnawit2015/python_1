```python
def karatsuba(x, y):
    x = str(x)
    y = str(y)
    
    max_len = max(len(x), len(y))
    if max_len == 1:
        return int(x) * int(y) 
    
    if len(x) < max_len:
        x = x.zfill(max_len)
    if len(y) < max_len:
        y = y.zfill(max_len)
        
    mid = max_len // 2
    x1, x0 = x[:-mid], x[-mid:]
    y1, y0 = y[:-mid], y[-mid:]
    x1, x0 = int(x1), int(x0)
    y1, y0 = int(y1), int(y0)
    
    
    z2 = karatsuba(x1, y1)
    z0 = karatsuba(x0, y0)
    z1 = karatsuba(x1 + x0, y1 + y0) - z2 - z0
    
    return (z2 * 10 ** (2 * mid)) + (z1 * 10 ** mid) + z0


print(karatsuba(23, 14))
print(karatsuba(12345678, 87654321))
print(karatsuba(9999999999999999, 88888888888888888888))
```

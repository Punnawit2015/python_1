```python
from itertools import combinations

def knapsak_brute_force_trace(weights, values, capity):
    n = len(weights)
    max_value = 0
    best_combination = []
    
    print("find all combination:")
    
    for r in range(1, n+1):
        for combo in combinations(range(n), r):
            total_weight = sum(weights[i] for i in combo)
            total_value = sum(values[i] for i in combo)
            
            
            #print trace
            print(f"Items {combo}: total_weight = {total_weight}, total_value = {total_value}", end='')
            
            if total_weight <= capity:
                if total_value > max_value:
                    max_value = total_value
                    best_combination = [combo]
                    print("------>NEW MAX!!!!!")
                elif total_value == max_value:
                    best_combination.append(combo)
                    print("------>MATCH MAX!")
                    
                else:
                    print()
            else:
                print("-----> EXCEEDS CAPITY!!!!!!")
                
    return max_value, best_combination
    
    
weights = [7,3,4,5]
values = [42,12,40,25]
capity = 10

max_val, combos = knapsak_brute_force_trace(weights, values, capity)
print("\nMax value:", max_val)
print("all best combinations:", combos)
 
```

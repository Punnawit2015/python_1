```python
import matplotlib.pyplot as plt 
```

```python
subject = ['Math', 'science', 'Thai', 'English']
Point = [25 , 20 , 30 , 15]
plt.bar(subject,Point,color = 'green')
plt.xlabel("Subject")
plt.ylabel("Point")
plt.title("Student in each subject")
plt.show()
```

```python
TotalSellary = ['Iphone 14' , 'Iphone 14 pro' , 'Iphone 15' , 'Iphone 16 pro max' , 'samsung galaxy S25']
Totalbuy = [2500 , 1000 , 2510 , 4000 , 3500]
plt.bar(TotalSellary,Totalbuy,color = 'Blue')
plt.xlabel("TotalSellary")
plt.ylabel("TotalBuy")
plt.title("Buy and Sellery")
plt.show() 
```

```python
RainingPerMonths = ['January' , 'February' , 'March' , 'April' , 'May']
TotalRain = [10 , 5 , 20 , 45 , 60]
plt.plot(RainingPerMonths, TotalRain,marker = 'o',linestyle = '-' ,color = 'blue')
plt.xlabel("RainingPerMonths")
plt.ylabel("TotalRain")
plt.title("Raining")
plt.grid()
plt.show() 
```

```python
SELLING = ['Iphone 14' , 'Iphone 14 pro' , 'Iphone 15' , 'Iphone 16 pro' , 'Samsung galaxy S25']
HowManySell = [2500 , 1000 , 2510 , 4000 , 3500]
plt.plot(SELLING, HowManySell,marker = 'o',linestyle = '-' ,color = 'green')
plt.xlabel("SELLING")
plt.ylabel("HowManySell")
plt.title("SELLING Phone")
plt.grid()
plt.show() 
```

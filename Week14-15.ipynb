```python
import matplotlib.pyplot as plt 
```

```python
subject = ['Math', 'science', 'Thai', 'English']
Point = [25 , 20 , 30 , 15]
plt.bar(subject,Point,color = 'green')
plt.xlabel("Subject")
plt.ylabel("Point")
plt.title("Student in each subject")
plt.show()
```

```python
RainingPerMonths = ['January' , 'February' , 'March' , 'April' , 'May']
TotalRain = [10 , 5 , 20 , 45 , 60]
plt.plot(RainingPerMonths, TotalRain,marker = 'o',linestyle = '-' ,color = 'blue')
plt.xlabel("RainingPerMonths")
plt.ylabel("TotalRain")
plt.title("Raining")
plt.grid()
plt.show() 
```

```python
RedingBook = ['NUT', 'BITHONG', 'OM' , 'JUM' ,'MEW']
ReadingBookPoint = [8 , 5 , 11 , 7 , 9]
plt.bar(RedingBook, ReadingBookPoint, color = 'purple')
plt.xlabel("RedingBook")
plt.ylabel("ReadingBookPoint")
plt.title("ReadingBookmark")

Mostbooks = max(ReadingBookPoint)
MostRead = RedingBook[ReadingBookPoint.index(Mostbooks)]
plusBook = sum(ReadingBookPoint)
print(f"Total points from reading books: {plusBook}")
print(f"Most read book is {MostRead} with {Mostbooks} points.")
plt.show()
```

```python
Days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
People = [28 , 25 , 30 , 27 , 24]
plt.bar(Days, People, color = 'blue')
plt.xlabel("Days")
plt.ylabel("People")
MostPeople = max(People)
MostDays = Days[People.index(MostPeople)]
lessPeople = min(People)
lessDays = Days[People.index(lessPeople)]
print(f"Least people on a day is {lessPeople} on {lessDays}.")
print(f"Most people on a day is {MostPeople} on {MostDays}.")
plt.grid()
plt.show()
```

```python
LikeToEatInLunch = ['SinglePlate', 'OrderDish', 'BreakFood', 'Dessert' , 'Other']
HowManyPeople = [35 , 25 , 20 , 10 , 10]
plt.bar(LikeToEatInLunch, HowManyPeople, color = 'LightBlue')
plt.xlabel("LikeToEatInLunch")
plt.ylabel("HowManyPeople")
plt.title("LikeToEatInLunch")
fav_break_count_ByPuNN  = HowManyPeople[LikeToEatInLunch.index('BreakFood')]
print(f"Most people like to eat in lunch is {fav_break_count_ByPuNN} people.")
max_people = max(LikeToEatInLunch)
max_people_count = HowManyPeople[LikeToEatInLunch.index(max_people)]
print(f"Most people like to eat in lunch is {max_people} with {max_people_count} people.")
plt.grid
plt.show
```

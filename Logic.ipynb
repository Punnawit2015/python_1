```python
# river_crossing.py
import math

# use sets and all-lowercase names
LEFT  = {"farmer", "wolf", "goat", "cabbage"}
RIGHT = set()

def print_state(step):
    print(f"Step {step}:")
    print("Left Bank:", sorted(LEFT))
    print("Right Bank:", sorted(RIGHT))
    print()

def move(item=None):
    """Move the farmer, optionally taking one item, across the river."""
    global LEFT, RIGHT

    # decide source and destination based on where the farmer is
    if "farmer" in LEFT:
        source, dest = LEFT, RIGHT
    else:
        source, dest = RIGHT, LEFT

    # move farmer
    source.remove("farmer")
    dest.add("farmer")

    # move item if specified
    if item:
        if item not in source:
            raise ValueError(f"Cannot move {item}: not on the same bank as the farmer.")
        source.remove(item)
        dest.add(item)

    # check safety after move
    if not check_valid():
        raise RuntimeError("Unsafe configuration! Wolf eats goat or goat eats cabbage.")

def is_safe(bank):
    """Return False if goat and cabbage or wolf and goat are left alone."""
    if "farmer" not in bank:
        if "wolf" in bank and "goat" in bank:
            return False
        if "goat" in bank and "cabbage" in bank:
            return False
    return True

def check_valid():
    """Both banks must be safe."""
    return is_safe(LEFT) and is_safe(RIGHT)

step = 0
print_state(step)

step += 1
move("goat")
print_state(step)

step += 1
move("")
print_state(step)

step += 1
move("cabbage")
print_state(step)

step += 1
move("goat")
print_state(step)

step += 1
move("wolf")
print_state(step)

step += 1
move("")
print_state(step)

step += 1
move("goat")
print_state(step)
```
